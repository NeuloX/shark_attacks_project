```python
import os
import pandas as pd
import re
from pdfminer.high_level import extract_text
```

```python
shark_data = pd.read_csv("clean_shark_dataV3.csv")
shark_data = shark_data.fillna("Not Defined")
shark_data.columns
```

```python
shark_data
```

```python
activity = shark_data["activity"].tolist()
act_list = []
for act in activity:
    act = act.lower()
    act_list.append(act)
shark_data["activity"] = act_list
shark_data["activity"].value_counts()
```

```python
clean_act = []
for acts in act_list:
    if acts.find("surf") != -1:
        acts = "surfing"
        clean_act.append(acts)
        continue
    elif acts.find("spearfishing") != -1:
        acts = "spear fishing"
        clean_act.append(acts)
        continue
    elif acts.find("fishing") != -1:
        acts = "fishing"
        clean_act.append(acts)
    elif acts.find("diving") != -1:
        acts = "diving"
        clean_act.append(acts)
    elif acts.find("swimming") != -1:
        acts = "swimming"
        clean_act.append(acts)
    elif acts.find("standing") != -1:
        acts = "standing"
        clean_act.append(acts)
    elif acts.find("paddleboarding") != -1 or acts.find("paddle boarding") != -1 or acts.find("paddle-boarding") != -1:
        acts = "paddle-boarding"
        clean_act.append(acts)
    elif acts.find("playing") != -1:
        acts = "playing"
        clean_act.append(acts)
    elif acts.find("wading") != -1:
        acts = "wading"
        clean_act.append(acts)
    elif acts.find("rowing") != -1:
        acts = "rowing"
        clean_act.append(acts)
    elif acts.find("kite") != -1:
        acts = "kite-surfing"
        clean_act.append(acts)
    elif acts.find("jump") != -1:
        acts = "jumped into the water"
        clean_act.append(acts)
    elif acts.find("sail") != -1:
        acts = "sailing"
        clean_act.append(acts)
    else:
        clean_act.append(acts)
shark_data["activity"] = clean_act
shark_data["activity"].unique()
```
